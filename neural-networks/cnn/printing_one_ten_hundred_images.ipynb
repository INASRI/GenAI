```python
from matplotlib import pyplot as plt
import numpy as np
import tensorflow as tf

# Load the MNIST dataset using Keras
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Plot the first 10 images from the test set
plt.figure(figsize=(10, 2))
for i in range(10):
    pixels = x_test[i].reshape((28, 28))
    plt.subplot(1, 10, i + 1)
    plt.imshow(pixels, cmap='gray')
    plt.title(str(y_test[i]), fontsize=10)
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
import tensorflow as tf

# Load the MNIST dataset using Keras
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Plot the first 100 images from the test set in a 10x10 grid
plt.figure(figsize=(10, 10))
for i in range(100):
    plt.subplot(10, 10, i + 1)
    pixels = x_test[i].reshape((28, 28))
    plt.imshow(pixels, cmap='gray')
    plt.title(str(y_test[i]), fontsize=8)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
