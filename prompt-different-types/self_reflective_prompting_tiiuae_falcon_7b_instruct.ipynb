```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
# --- Load an Open Source Model ---
# (A small and fast open model: Mistral-7B-Instruct, Llama-2-7B, etc.)
# For demo, using a small model from HuggingFace
model_name = "tiiuae/falcon-7b-instruct"  # or "facebook/opt-1.3b" if lower resources
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16, device_map="auto")
```

```python
# --- Helper Function to Generate Text ---
def generate_response(prompt, max_new_tokens=200):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    outputs = model.generate(**inputs, max_new_tokens=max_new_tokens, do_sample=True)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
# --- 1. Ask initial question ---
user_question = "Explain why the sky is blue in simple terms."

# Step 1: Generate initial answer
initial_prompt = f"Answer the following question:\n\n{user_question}\n\nAnswer:"
initial_answer = generate_response(initial_prompt)

print("\n🧠 Initial Answer:\n", initial_answer)
```

```python
# --- 2. Ask for Reflection ---
reflection_prompt = f"Review the following answer critically and suggest improvements if needed:\n\n{initial_answer}\n\nReflection:"
reflection = generate_response(reflection_prompt)

print("\n🔍 Reflection on Initial Answer:\n", reflection)

# --- 3. Optional: Regenerate based on Reflection (simplified) ---
improvement_prompt = f"Based on the following reflection, improve the original answer:\n\nReflection: {reflection}\n\nImproved Answer:"
improved_answer = generate_response(improvement_prompt)

print("\n🚀 Improved Answer:\n", improved_answer)
```
