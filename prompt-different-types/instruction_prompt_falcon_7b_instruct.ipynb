```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
# --- Step 1: Load the Open-Source Model ---
# You can replace this with another model you like
model_name = "tiiuae/falcon-7b-instruct"  # Example: Falcon-7B Instruct
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16, device_map="auto")
```

```python
# --- Step 2: Write a Simple Instruction Prompt ---
instruction = """Summarize the following paragraph:

Paragraph: Artificial intelligence is a branch of computer science that aims to create machines capable of intelligent behavior, such as learning, reasoning, and problem-solving.

Summary:"""
```

```python
# --- Step 3: Tokenize and Generate Output ---
inputs = tokenizer(instruction, return_tensors="pt").to(model.device)

# Generate
outputs = model.generate(
    **inputs,
    max_new_tokens=150,
    do_sample=True,   # Optional: sampling for creativity
    temperature=0.7   # Optional: controls randomness
)
```

```python
# --- Step 4: Decode and Print ---
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("\n🧠 Model Output:\n", generated_text)
```
