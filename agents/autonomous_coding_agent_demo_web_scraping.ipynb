```python
import os
from openai import OpenAI
from dotenv import load_dotenv
```

```python
# Load API key
load_dotenv()
client = OpenAI(api_key="")


def ask_gpt(messages, model="gpt-4", temperature=0.3):
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature
    )
    return response.choices[0].message.content


def breakdown_task(task):
    system_msg = {"role": "system", "content": "You are a senior software engineer breaking down coding problems into steps."}
    user_msg = {"role": "user", "content": f"Break this task into sub-steps: {task}"}
    return ask_gpt([system_msg, user_msg])


def generate_code_for_step(step):
    system_msg = {"role": "system", "content": "You are a Python expert. Generate Python code only. No explanations."}
    user_msg = {"role": "user", "content": f"Write Python code for: {step}"}
    return ask_gpt([system_msg, user_msg])


def debug_code(code, error):
    system_msg = {"role": "system", "content": "You are a Python debugger. Fix the code."}
    user_msg = {"role": "user", "content": f"This code:\n{code}\ncaused this error:\n{error}\nFix and return code only."}
    return ask_gpt([system_msg, user_msg])


def execute_code(code):
    try:
        exec_globals = {}
        exec(code, exec_globals)
        return "‚úÖ Executed successfully"
    except Exception as e:
        return f"‚ùå Error during execution: {e}"


def autonomous_coding_agent(task):
    print(f"\nüîç Task: {task}")
    print("\nüìã Breaking into subtasks...\n")
    subtasks = breakdown_task(task)
    print(subtasks)

    steps = subtasks.strip().split("\n")
    final_code = ""
    for step in steps:
        if not step.strip():
            continue
        print(f"\nüõ† Working on: {step}")
        code = generate_code_for_step(step)
        print("üëá Generated Code:\n", code)

        result = execute_code(code)
        print(result)

        if "Error" in result:
            code = debug_code(code, result)
            print("\nüîÅ Retrying with fixed code...\n", code)
            result = execute_code(code)
            print(result)

        final_code += f"\n# Step: {step}\n{code}\n"

    return final_code


if __name__ == "__main__":
    user_task = input("üß† What should the coding agent build?\n> ")
    output_code = autonomous_coding_agent(user_task)

    with open("final_output.py", "w") as f:
        f.write(output_code)

    print("\n‚úÖ Done! Final code saved to `final_output.py`.")
```
